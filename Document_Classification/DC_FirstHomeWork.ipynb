[Свертка списка](https://ru.wikipedia.org/wiki/%D0%A1%D0%B2%D1%91%D1%80%D1%82%D0%BA%D0%B0_%D1%81%D0%BF%D0%B8%D1%81%D0%BA%D0%B0) -  это обобщенная операция над списком, c помощью которой можно преобразовать список в единое значение. Например, рассмотрим реализации свертки слева и свертки справа (левоассоциативную свертку и правоассоциативную свертку):

```python
def foldl(f, x0, lst):
    if not lst:
        return x0
    return foldl(f, f(x0, lst[0]), lst[1:])

def foldr(f, x0, lst):
    if not lst:
        return x0
    return f(lst[0], foldr(f, x0, lst[1:]))
```

```python
#сумма
foldl(lambda x, y: x + y, 0, [1, 2, 3])
```

```python
#конкатенация
foldl(lambda x, y: '{}{}'.format(x, y), '', [1, 2, 3])
```

# Задание 2

нужно написать функцию, которая принимает две строки и проверяет, входит ли хотя бы одна перестановка второй строки в первую. Например:

> a = 'abcrotm'
> 
> b = 'tro'

функция def check_inv(a, b) вернет True, так как 'rot' содержится в 'abcrotm'. Нужно подумать как можно более оптимальный алгоритм и оценить его сложность. 

```python
def check_inv(a, b):
    startDict = {b[i]: 0 for i in range(len(b))}
    for elB in b:
        startDict[elB] += 1
    finishDict = startDict.copy()
    for elA in a:
        if (elA in b):
            if (finishDict[elA] > 0): 
                finishDict[elA] -= 1
            else:
                finishDict = startDict.copy()
                finishDict[elA] -= 1
        else: 
            finishDict = startDict.copy()
        s = 0
        for value in finishDict.values():
            s += value
        if (s == 0):
            return True
        
    return False


print(check_inv('aabb','abb'))
print(check_inv('baaaaaaaaaaaaaac','abc'))
print(check_inv('asdfqwertygfsddfg', 'tryewq'))
print(check_inv('rewqffff', 'ffqwerf'))
```


# Задание 3

Реализовать бинарное дерево (класс Tree), в нём методы __repr__, __str__, __iter__ (итерация только по листьям).

```python
class Tree:
    def __init__(self, value=None, left=None, right=None):
        self.left = left
        self.right = right
        self.value = value
        
        
    def __iter__(self):
        if (self.right is None) and (self.left is None):
            yield self.value
        else:
            yield from self.left
            yield from self.right

    
    def __repr__(self):
        if (self.right is None and self.left is None):
            return 'Tree(' + repr(self.value) + ')'
        return 'Tree(' + repr(self.value) + ', ' + repr(self.left) + ', ' + repr(self.right) + ')'
    
    
    def __str__(self, level = 0):
        r = "\t" * level + str(self.value) + "\n"
        if(self.left is not None):
            r += self.left.__str__(level+1)
        if(self.right is not None):
            r += self.right.__str__(level+1)
        return r
    
tree = Tree(0, Tree(1, Tree(3), Tree(4)),                             
               Tree(2))

print(str(tree))
```
